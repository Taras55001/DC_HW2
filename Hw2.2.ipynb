```python
#Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
%matplotlib inline
import pandas as pd
df = pd.read_csv('./2017_jun_final.csv')
```

```python
#Визначте розмір таблиці за допомогою методу shape
print(f"кількість рядків та стовпців:{df.shape}")
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(df.dtypes)
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Частка пропусків в кожній колонці:")
print(missing_percentage)
```

```python
#Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
columns_with_na = df.columns[df.isna().any()]
columns_to_keep = ['Язык.программирования']
columns_to_drop = [col for col in columns_with_na if col not in columns_to_keep]
df = df.drop(columns=columns_to_drop)
```

```python
#Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Частка пропусків в кожній колонці:")
print(missing_percentage)
```

```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df.dropna(inplace=True)
```

```python
#Визначте новий розмір таблиці за допомогою методу shape
new_shape = df.shape
print("Новий розмір таблиці після видалення рядків з пропусками:", new_shape)
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = df[df['Язык.программирования'].str.contains('Python', case=False, na=False)]
python_data.head()
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape
print("Новий розмір таблиці python_data:", python_data.shape)
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped_by_position = df.groupby('Должность')
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада",
#виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
aggregated_data = grouped_by_position.agg({'Зарплата.в.месяц': ['min', 'max']})
aggregated_data.columns = ['Мінімальна_зарплата', 'Максимальна_зарплата']

print(aggregated_data)
```

```python
import matplotlib.pyplot as plt

min_salary = aggregated_data['Мінімальна_зарплата']
max_salary = aggregated_data['Максимальна_зарплата']

max_height = max(7, 0)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.hist(min_salary, bins=15, edgecolor='k', alpha=0.7, color='#f748ce')
plt.title('Гістограма мінімальних зарплат')
plt.xlabel('Мінімальна зарплата')
plt.ylabel('Кількість посад')
plt.ylim(0, max_height)

plt.subplot(1, 2, 2)
plt.hist(max_salary, bins=15, edgecolor='k', alpha=0.7, color='#1df0d0')
plt.title('Гістограма максимальних зарплат')
plt.xlabel('Максимальна зарплата')
plt.ylabel('Кількість посад')
plt.ylim(0, max_height)

plt.tight_layout()
plt.show()
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
#Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    avg_salary = row['Зарплата.в.месяц'] + row['Изменение.зарплаты.за.12.месяцев'] / 12
    row['avg'] = avg_salary
    return row

python_data = python_data.apply(fill_avg_salary, axis=1)
python_data.head()
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
avg_salary_stats = python_data['avg'].describe()
print(avg_salary_stats)
```

```python
#Збережіть отриману таблицю в CSV файл
python_data.to_csv('./new_data_file.csv', index=False)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(40, 20))
ax = fig.add_subplot(projection='3d')

x = df['Должность']
y = df['Зарплата.в.месяц'].astype(float)

df['Общий.опыт.работы'] = df['Общий.опыт.работы'].replace('10 и более лет', '10')
df['Общий.опыт.работы'] = df['Общий.опыт.работы'].replace('меньше 3 месяцев', '0')
df['Общий.опыт.работы'] = pd.to_numeric(df['Общий.опыт.работы'])
z = df['Общий.опыт.работы'].astype(float)

positions = df['Должность'].unique()
position_ids = range(len(positions))
colors = {'Data Scientist / Data Analyst': 'r', 'DevOps': 'b', 'ERP / CRM': 'g', 'HTML coder': 'y',
          'Junior Software Engineer': 'c', 'Product Manager': 'm', 'Senior Software Engineer': 'y',
          'Software Engineer': 'k', 'System Architect': 'b', 'Team lead': 'g', 'Technical Lead': 'r'}

size = 40

for position_id, position in zip(position_ids, positions):
    subset = df[df['Должность'] == position]
    ax.scatter([position_id] * len(subset), subset['Зарплата.в.месяц'], subset['Общий.опыт.работы'],
               c=colors[position], s=size, label=position)

ax.set_xlabel('Посада', color='midnightblue')
ax.set_xticks(position_ids)
ax.set_xticklabels(positions, rotation=45, ha='center', fontsize=10)
ax.set_ylabel('Зарплата.в.месяц', color='midnightblue')
ax.set_zlabel('Общий.опыт.работы', color='midnightblue')
ax.legend(loc='upper right', prop={'size': 12})
plt.show()
```
